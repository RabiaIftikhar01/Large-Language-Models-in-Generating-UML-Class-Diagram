```python
import json
from IPython.display import display, Markdown, HTML
from openai import OpenAI

client = OpenAI(api_key=openai_api_key)
GPT_MODEL = 'gpt-4o-mini'
```

```python
f = open("Good Design Practices.txt", "r")
practices = f.read() 
f.close()

f = open("Instructions.txt", "r")
instructions = f.read() 
f.close()

import json

# Open and read the JSON file
with open('dataset.json', 'r') as file:
    dataset = json.load(file)
dataset[0]
```

```python
prompt=f"""You are a professional Software engineer with years of experience 
in Software analysis and design using UML. For the generation task below, you need 
to generate class diagram in plantUML format that fulfills all the design practices 
and instructions provided below:
Design practice: {practices}
Instructions: {instructions}

Besides, I am also providing you some examples on how to complete this task. All the examples
consist of a generation task, two possible UML solutions to that task, and a choice representing a preferable solution.
Train yourself on these examples before completing the given generation task.

********************************************
Example 1:

generation_task: {dataset[0]['generation_task']}
solution_1: {dataset[0]['solution_1']}
solution_2: {dataset[0]['solution_2']}
choice: {dataset[0]['choice']}

********************************************
Example 2:

generation_task: {dataset[1]['generation_task']}
solution_1: {dataset[1]['solution_1']}
solution_2: {dataset[1]['solution_2']}
choice: {dataset[1]['choice']}
********************************************
Example 3:

generation_task: {dataset[2]['generation_task']}
solution_1: {dataset[2]['solution_1']}
solution_2: {dataset[2]['solution_2']}
choice: {dataset[2]['choice']}
********************************************
Example 4:

generation_task: {dataset[3]['generation_task']}
solution_1: {dataset[3]['solution_1']}
solution_2: {dataset[3]['solution_2']}
choice: {dataset[3]['choice']}
********************************************
Example 5:

generation_task: {dataset[4]['generation_task']}
solution_1: {dataset[4]['solution_1']}
solution_2: {dataset[4]['solution_2']}
choice: {dataset[4]['choice']}
********************************************


Your Task:
A university wants to extend its learning platform. 
Users can take part in courses in multiple capacities 
(i.e. Student, and Instructor). Only instructors are 
allowed to publish new course offerings, which may 
differ by format (i.e. Undergraduate, Postgraduate), 
and update course content. Enrollments must keep track 
of the enrollment date and current status 
(e.g., waitlisted, active, completed). Students should 
be able to access course material but not modify it. 
Course participants and instructors should receive 
notifications (e.g., assignment deadlines, announcements, 
assignment submitted).Provide the output in plantUML format. 
Provide only diagram no explaintion needed

Your solution:
"""

```

```python
response = client.chat.completions.create(model=GPT_MODEL,messages=[{"role":"user","content": prompt}])
display(HTML('<div style="background-color: #f0fff8; padding: 10px; border-radius: 5px; border: 1px solid #d3d3d3;"></hr><h2>ðŸ”½ &nbsp; Markdown Output â€“ Beginning</h2></hr></div>'))
display(Markdown(response.choices[0].message.content))
display(HTML('<div style="background-color: #fff4f4; padding: 10px; border-radius: 5px; border: 1px solid #d3d3d3;"></hr><h2>ðŸ”¼ &nbsp; Markdown Output â€“ End</h2></hr></div>'))
```
